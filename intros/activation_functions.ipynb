```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(val):
    return 1 / (1 + np.exp(-val))

# plot sigmoid for values from -10 to 10
vals=np.linspace(-10, 10, 100)
plt.plot(vals, sigmoid(vals))
plt.show()
    
```

```python
def relu(val):
    return np.maximum(0, val)


# plot relu for values from -10 to 10
vals=np.linspace(-10, 10, 100)
plt.plot(vals, relu(vals))
plt.show()
```
